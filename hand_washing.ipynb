```python
def extract_country_row(df, country_name):
    df.loc[df['Country Name'] == country_name] = df.loc[df['Country Name'] == country_name].fillna(0)
    return df.loc[df['Country Name'] == country_name]

def plot_country_data(df, country_name, plot_line):
    starting_year = 2000    
    y = np.array(extract_country_row(df, country_name).iloc[:, -18:]).flatten()
    for i in range(len(y)):
        if y[i] != 0:
            break
            
    starting_year += i
    y = y[i:]
    x = np.array(range(starting_year, 2018))
    
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, y, 'o')
    if plot_line:
        plt.plot(x, m*x + b)
        print(m)
        print(b)
    plt.title('% of population of ' + country_name + ' with access to HWFs vs Time')
    plt.xlabel('Time (years)')
    plt.ylabel('% of population with access to HWFs')
    
    plt.savefig(country_name + 'data.png')
```

```python
plot_country_data(df, "Liberia", True)
```

```python
plot_country_data(df, "Turkmenistan", True)
```

```python
plot_country_data(df, "Armenia", False)
```
